```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 5000, 7000, 8000, 9000],
    [2, 11000, 13000, 19000, 29000],
    [3, 20000, 23000, 26000, 31000],
    [4, 18000, 21000, 24000, 27000],
    [5, 39000, 42000, 46000, 50000]
])

print("=== Zomato sales  analysys ===")
print("\n Sample data for 1st restaurant:", sales_data[:3])
print("\n Sample data for 1st restaurant:", sales_data[:, 1:])
```

```python
# total sales per year 
print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)


# Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis =1)
print(cumsum)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis =0))
plt.title("Average sales over years")
plt.xlabel("Years")
plt.ylabel("Average sales")
plt.grid(True)
plt.show()
```
